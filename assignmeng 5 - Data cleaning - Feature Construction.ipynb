# 1 Import Libiraries

```python
import pandas as pd
import numpy as np
import gc
from sklearn.preprocessing import LabelEncoder

train = pd.read_csv('data/train_after_feature_selection.csv')
test = pd.read_csv('data/test_after_feature_selection.csv')
```

# 2 Feature Construction

## 2.1 Transaction Amount

### 2.1.1 TransactionAmt into bin

```python
train['TransactionAmt_bin'] = pd.cut(train['TransactionAmt'].astype(int),500)
test['TransactionAmt_bin'] = pd.cut(test['TransactionAmt'].astype(int),500)

train.head(20)
```

### 2.1.2 TransactionAmt_bin encoding

```python
label = LabelEncoder()
train['TransactionAmt_bin_code'] = label.fit_transform(train['TransactionAmt_bin'])
test['TransactionAmt_bin_code'] = label.fit_transform(test['TransactionAmt_bin'])
train.head(10)
```

## 2.2 card 1

### 2.2.1 card1 into bin

```python
train['card1_bin'] = pd.cut(train['card1'].astype(int),20)
test['card1_bin'] = pd.cut(test['card1'].astype(int),20)
train.head(10)
```

### 2.2.2 card1_bin encoding

```python
label = LabelEncoder()
train['card1_bin_code'] = label.fit_transform(train['card1_bin'])
test['card1_bin_code'] = label.fit_transform(test['card1_bin'])
train.head(10)
```

## 2.3 Card 2

### 2.3.1 Card1 Into Bin

```python
train['card2_bin'] = pd.cut(train['card2'].astype(int),20)
test['card2_bin'] = pd.cut(test['card2'].astype(int),20)
train.head(10)
```

### 2.3.2 Card2 Encoding

```python
label = LabelEncoder()
train['card2_bin_code'] = label.fit_transform(train['card2_bin'])
test['card2_bin_code'] = label.fit_transform(test['card2_bin'])
train.head(10)
```

### 2.4.1 Distance Into Bin

```python
train['dist1_bin'] = pd.cut(train['dist1'].astype(int),500)
test['dist1_bin'] = pd.cut(test['dist1'].astype(int),500)
train.head(10)
```

### 2.4.2 Distance Encoding

```python
label = LabelEncoder()
train['dist1_bin_code'] = label.fit_transform(train['dist1_bin'])
test['dist1_bin_code'] = label.fit_transform(test['dist1_bin'])
train.head(10)
```
